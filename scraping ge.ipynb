```python
import re
import json
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
url2 = 'https://ge.globo.com/futebol/brasileirao-serie-a/' 
content = requests.get(url2)
print(content.status_code)

soup = bs(content.text, 'html')
tag = soup.findAll('script', attrs = {'id' : 'scriptReact'})[0].string
tabela = re.findall('const classificacao = (.*?);', tag)[0]
```

```python
dic = {}
characteristics = ['ordem', 'nome_popular', 'pontos', 'jogos',
                   'vitorias', 'empates', 'derrotas', 'gols_pro',
                   'gols_contra', 'saldo_gols', 'aproveitamento', 'ultimos_jogos']

for i in range(20):
    if i == 0:
        begin = tabela.find('{', 1)
        end = tabela.find('}', tabela.find('}') + 1) + 1
        data = tabela[begin:end]
    else:
        begin = tabela.find('{', end)
        end = tabela.find('}', tabela.find('}', begin) + 1) + 1
        data = tabela[begin:end]
        
    data = json.loads(data)

    for characteristic in characteristics:        
        if i == 0:
            dic[characteristic] = []
            
        dic[characteristic].append(data[characteristic])
        
df = pd.DataFrame.from_dict(dic, orient = 'columns', dtype = None)
df.set_index('ordem', inplace = True)
df
```

```python
tabela = json.loads(tabela)
dic = {}
characteristics = ['ordem', 'nome_popular', 'pontos', 'jogos',
                   'vitorias', 'empates', 'derrotas', 'gols_pro',
                   'gols_contra', 'saldo_gols', 'aproveitamento', 'ultimos_jogos']

for characteristic in characteristics:
    dic[characteristic] = []

for club in tabela['classificacao']:
    for characteristic in characteristics:            
        dic[characteristic].append(club[characteristic])
        
df = pd.DataFrame.from_dict(dic, orient = 'columns', dtype = None)
df.set_index('ordem', inplace = True)
df
```
